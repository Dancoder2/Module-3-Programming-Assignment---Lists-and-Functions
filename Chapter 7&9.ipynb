9.2
Define a generator function called get_odds() that yields
the odd numbers from range(10). Print the third one returned.


```python
def get_odds():
    for number in range(10):
        if number % 2 == 1:
            yield number

count = 0
for odd in get_odds():
    count += 1
    if count == 3:
        print(odd)
        break
```
